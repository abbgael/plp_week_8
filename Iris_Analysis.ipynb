# Iris Dataset Analysis and Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Load the dataset
iris = load_iris(as_frame=True)
df = iris.frame
```

## Task 3: Visualizations

```python
# Line Chart
df['Index'] = df.index
plt.figure(figsize=(10, 6))
plt.plot(df['Index'], df['sepal length (cm)'], label='Sepal Length')
plt.plot(df['Index'], df['petal length (cm)'], label='Petal Length')
plt.title('Line Chart: Sepal and Petal Length Over Index')
plt.xlabel('Index')
plt.ylabel('Length (cm)')
plt.legend()
plt.show()
```

```python
# Bar Chart - Updated for Seaborn 0.12+
plt.figure(figsize=(8, 5))
sns.barplot(x='target', y='petal length (cm)', data=df, errorbar=None)
plt.title('Average Petal Length per Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Length (cm)')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 5))
plt.hist(df['sepal width (cm)'], bins=15, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Width')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', hue='target', data=df)
plt.title('Scatter Plot: Sepal Length vs Petal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
